```python
#check if there is any missing data
data.isna().sum()
```

```python
#ny viewing the data we figured out that column "Unamed: 0 " is not useful for us so we must drop it
data.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
#ny viewing the data we figured out that column "id " is not useful for us so we must drop it

data.drop('id', axis=1, inplace=True)
```

```python
#checking after
data.isna().sum()
```

```python
#to know the number if classes for the data
data['diagnosis'].unique()
```

```python
# M (Malignant) = Bad, B (Benin) = Good
data['diagnosis'] = data['diagnosis'].map({'M':0,'B':1})
```

```python
# data is class_data
# pop vs drop
# What parameters to delete, mean or worst?

data.drop(['perimeter_mean', 'area_mean'], axis=1, inplace=True)
data.drop(['perimeter_worst', 'area_worst', 'radius_worst', 'texture_worst'], axis=1, inplace=True)
data.drop(['perimeter_se', 'area_se'], axis=1, inplace=True)
data.drop(['compactness_worst', 'concavity_worst', 'concave points_worst', 'smoothness_mean', 'concavity_mean', 'concave points_mean'], axis=1, inplace=True)
data.drop(['concavity_se', 'concave points_se'], axis=1, inplace=True)
data.drop(['radius_se', 'fractal_dimension_worst', 'symmetry_worst', 'fractal_dimension_worst', 'compactness_se', 'fractal_dimension_mean'], axis=1, inplace=True)
```

```python
plt.figure(figsize=(18,13))
sns.heatmap(data.corr())
```

```python
#spliting the output (target) from the feature
target = data.pop('diagnosis')
data.head(10)
```

```python
# training svc model using our data
from sklearn import svm
from sklearn.model_selection import cross_val_score
clf = svm.SVC()
score = cross_val_score(clf, data, target, cv=5)
np.mean(score)
```

```python
#hyper tunning to find the best parameters and accuracy
from sklearn.model_selection import GridSearchCV
clf = svm.SVC()
param_grid ={'kernel':["linear","rbf"],"C" : [1, 5 , 10],'gamma':[0.1 , 0.5 ,0.8]}
grdCV = GridSearchCV(clf,param_grid,cv=5)
grdCV.fit(data, target)
```

```python
grdCV.best_params_
```

```python
grdCV.best_score_
```

```python
# scaling the data to avoid miss classfication
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(data)
```

```python
#trainig kneighbors model
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
# np.arrange => numpy list, faster than normal python list
param_grid ={'n_neighbors':np.arange(1,25,2)}
grdCV = GridSearchCV(clf,param_grid,cv=5)
grdCV.fit(scaled, target)
```

```python
grdCV.best_params_
```

```python
# Best one so far.
grdCV.best_score_
```
